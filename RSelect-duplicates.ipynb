```python
import random
import sys

# Choose a pivot
def choosePivot(n, pivot_type = 'random'):

    if pivot_type == 'random':
        index = random.randint(0, n-1)
    
    elif pivot_type == 'fixed':
        index = 0 # Default method chooses first element
    
    else:
        sys.exit("Error, unexpected pivot type used.")
    
    return index

# partion around the pivot using repeated swaps
def partPivot(a, pivot):

    #print(a, pivot)
    
    # Swap pivot with first element
    if pivot != 0:
        a[0], a[pivot] = a[pivot], a[0]
    
    #print(a)
    
    # Perform linear scan through array a
    # Initialise parition frontier
    i = 1
    k = 1 # number of elements equal to pivot element (including itself)
    
    for j in range(1, len(a)):
        if a[j] < a[0]:
            print("less:", a)
           # Swap scanned element with parition boundary
            a[i], a[j] = a[j], a[i] 
            i += 1
        if a[j] == a[0]:
            print("match:", a)
            # Move element matching pivot to front of array and circularly permute other elements, increase i by 1
            a[0:(j+1)] = [a[j]] + a[0:j]
            i += 1
            # Keep track of number of matching elements at front ('length' of pivot)
            k += 1
    
    # Put pivot at boundary
    # a[0], a[i-1] = a[i-1], a[0]
    # Need a version that moves all k matching elements at front to pivot boundary (including the elements themselves in i-1 returned)
    for l in range(k):
        a[l], a[i-1-l] = a[i-1-l], a[l]
    
    return a, i-1, k

def QuickSort(a, pivot_type = 'random'):
        
    n = len(a)
 
    # Base cases
#    if n == 0:
#        return []
    
    if n == 1:
        return a
    
    # Choose a pivot element
    pivot = choosePivot(n, pivot_type = pivot_type)
    
    # Partition array around pivot and return array and final position (ith order statistic) of pivot
    a, pivot_ios, width = partPivot(a, pivot) 
    
        
   # print(a, n, ios)
    
    #if ios == 0:
    #    return a
    
    # Recusively sort all elements less than pivot and greater than pivot
    if len(a[:pivot_ios]) > 0:
        a[:pivot_ios] = QuickSort(a[:pivot_ios], pivot_type = pivot_type)
    
    if len(a[(pivot_ios+1):]) > 0:
        a[(pivot_ios+1):] = QuickSort(a[(pivot_ios+1):], pivot_type = pivot_type)
    
    return a

# Define function to return ith order statistic of an array
def RSelect(a, i, pivot_type = 'random'):
    
    n = len(a)
    
    if i > n:
        sys.exit("ith order statistic exceeds array size")
    elif n == 1:
        return a
    else:
        
        # Choose pivot element
        pivot = choosePivot(n, pivot_type = pivot_type)

        print("Pre:", a, pivot)
        
        # Partition array around pivot and return array and final position (ith order statistic) of pivot
        a, pivot_ios, width = partPivot(a, pivot)

        print("Post:", a, pivot, pivot_ios, width)
        
        ## Narrow search based on results
        for l in range(width):
            if (pivot_ios - l) == (i - 1):
                return a[i - 1]
            
        if (pivot_ios - l) < (i-1):
            a = RSelect(a[(pivot_ios + 2 - l):], i - pivot_ios - 1)
        else:
            a = RSelect(a[:pivot_ios], i)
        
    return a
```
