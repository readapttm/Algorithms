```python
# Random contraction

import numpy as np
import random


## example contraction if removing (1, 2)
## Name vertices with integers

# output = [(1, 3), (1, 3), (1, 4)]
# i - identify second component in edge - in this case 2
# i - replace 2 with 1 in any other node in the edge list
# Delete 

def nodes(edge_list):
    
    nodes = np.concatenate((edge_list['source'], edge_list['sink']))
    nodes.sort()
    
    return np.unique(nodes)

def contract(edge_list, edge):
    
    # Identify selected edge
    edge_to_contract = edge_list[edge-1]
    
    # Identify node (sink) to be removed
    node_to_delete = edge_to_contract[1]
    
    # Remove edge
    edge_list = np.delete(edge_list, edge-1)
    
    edges = len(edge_list)
    
    # Loop through edges
    for i in range(edges):
       
        #print("edge_list:", edge_list, "i", i, "edge_list[i][0]:", edge_list[i][0], "edge_list[i][1]", edge_list[i][1])
        
                # Remove nodes
        if edge_list[i][1] == node_to_delete:
            edge_list[i][1] = edge_to_contract[0]
    
    # Remove self-edges created by removing nodes
    self_edges = np.array(edge_list['source'] == edge_list['sink'])
    edge_list = edge_list[~self_edges]
    #print("self_edges:", self_edges)
    
    return edge_list

def random_contract(edge_list):
    
    #edges = len(edge_list)
    #edge_counter = edges
    
    # Contract until only two nodes remain (although they could be connected by multiple edges)
    # Will need i - a while loop, ii - a count of the remaining nodes
    #for i in range(edges):
    while len(nodes(edge_list)) > 2:   
        edge_to_contract = random.randint(1, len(edge_list))
        #print(edge_to_contract)
        edge_list = contract(edge_list, edge_to_contract)
        #print(edge_list)

    return edge_list
    
# Create function to run random_contract 1000 times to develop min cut statistics    
def simulation(edge_list, n): # To fix
    
    results = []
    
    for i in range(n):
        results.append(np.array2string(random_contract(edge_list)))
        
    return results
    
# 1 Choose any edge (connecting two vertices)
# 2 Delete the edge
# - Vertices connected by edge merge
# - Update nodes on edge list to next node along
```
