Algorithm to find the closest pair of points in a 2D plane

```python
import numpy as np
import math
```

```python
# use a numpy array to make accessing X/Y dimension easy

dtype = [('X', 'float'), ('Y', 'float')]
coords = [(1, 2), (5, 3), (4, -1), (8, 2), (10, 2), (1, 1.9)]

ps = np.array(coords, dtype = dtype)

## Function to calculate Euclidean distance
def distance(p, q):
    return ((p['X'] - q['X'])**2 + (p['Y'] - q['Y'])**2)**0.5
```

```python
# Brute force algorithm
# iterate through all combinations i = range(n), j = range(i+1, n)



def closest_pair_bf(Px):
    
    points = len(Px)
    bestPair = (Px[0], Px[1])
    bestDist = distance(Px[0], Px[1])
    
    for i in range(1, points):
        for j in range(i+1, points):
            pairDist = distance(Px[i], Px[j])
            
            if pairDist < bestDist:
                bestPair = (Px[i], Px[j])
                bestDist = pairDist
                
    return bestPair, bestDist
```
